```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
from sklearn.datasets import make_blobs
```

```python
points, cluster_indexes = make_blobs(n_samples=300, centers=4,
cluster_std=0.8, random_state=0)
x = points[:, 0]
y = points[:, 1]
plt.scatter(x, y, s=50, alpha=0.7)
plt.show()
```

#### KMeans class Cluster

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=6, random_state=0) #! change the n_cluster number and see the result 
kmeans.fit(points)
predicted_cluster_indexes = kmeans.predict(points)

plt.scatter(x,y, c= predicted_cluster_indexes, s=50, alpha=0.7, cmap='viridis')
centers = kmeans.cluster_centers_

plt.scatter(centers[:,0], centers[:,1], c="red", s=100)
plt.show()
```

```python
# One way to pick the right number is with the elbow method,
# which plots inertias obtained from KMeans. inertia as a function of cluster counts

inertias = []

for i in range(1,10):
    kmeans = KMeans(n_clusters=i, random_state =0)
    kmeans.fit(points)
    inertias.append(kmeans.inertia_)
    
plt.plot(range(1,10), inertias)
plt.xlabel("Number of cluster")
plt.ylabel("Inertia")
# Plot a vertical line in min of inertis
plt.axvline(pd.Series(inertias).argmin(), color = "red")
plt.show()
```
